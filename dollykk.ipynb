```python
import textwrap
from instruct_pipeline import InstructionTextGenerationPipeline
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
```

```python
def print_response(response: str) -> str:
    print("\n".join(textwrap.wrap(response, width=80)))
```

```python
tokenizer = AutoTokenizer.from_pretrained(
    "databricks/dolly-v2-2-8b",
    padding_side="left"
)
model = AutoModelForCausalLM.from_pretrained(
    "databricks/dolly-v2-2-8b",
    torch_dtype=torch.bfloat16,
    device_map="auto"
)

generate_text = InstructionTextGenerationPipeline(model=model, tokenizer=tokenizer)
```

```python
print_response(
    generate_text("What is the meaning of life? Use no more than 3 sentences.")
)
```

```python
print_response(
    generate_text("I am feeling depressed. Give me necessary rescue session.")
)
```
